```python
import numpy as np
import pandas as pd

# Define a function to extract the country from the 'dir_pob' column
def extract_country(place):
    if pd.isnull(place):  # Check if the value is None or NaN
        return None
    else:
        # Split the place based on the comma and extract the last part (country)
        parts = place.split(', ')
        return parts[-1]

df = pd.read_csv('movie_data.csv')
df.columns = ['critic', 'crit_org', 'crit_cn', 'movie', 'm_year', 'director', 'dir_dob', 'dir_pob']
# Apply the function to create a new column 'cob'
df['cob'] = df['dir_pob'].apply(extract_country)
# Write the updated DataFrame to a new CSV file
df.to_csv('movie_data_updated.csv', index=False)
df.tail()
```

```python
def process_to_geojson(file):
    geo_data = {"type": "FeatureCollection", "features":[]}
    for row in file:
        this_dict = {"type": "Feature", "properties":{}, "geometry": {}}
        for key, value in row.items():
            key_names = key.split('.')
            if key_names[0] == 'geometry':
                this_dict['geometry'][key_names[1]] = value
            if str(key_names[0]) == 'properties':
                this_dict['properties'][key_names[1]] = value
        geo_data['features'].append(this_dict)
    return geo_data
```
